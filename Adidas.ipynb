# Adidas dataset analysing

```python
#impoting required libraries
from numpy import *
from pandas import *
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# reading the data
data=read_csv(r'C:\Users\DELL\Downloads\adidas.csv')
```

## Data preprocessing

```python
#looking at the data
data.head()
```

```python
data.info()
```

```python
data["original_price"]=data["original_price"].str.strip("$")
```

```python
data.info()
```

```python
#changing data types
data["original_price"]=data["original_price"].astype("float64")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#dropping missing values
data=data.dropna()
```

```python
data.info()
```

```python
#checking for duplicate values
data.duplicated().sum()
```

```python
data.head()
```

```python
#Creating a new column
data["Discount"]=data["original_price"]-data["selling_price"]
```

```python
data["Discount"].describe()
```

```python
data["crawled_at"]=to_datetime(data["crawled_at"])
```

```python
data.info()
```

```python
data["crawled_at"]
```

```python
data.columns
```

```python
#dropping columns
data=data.drop(columns=["url","source_website","images"])
```

```python
data.columns
```

```python
data.head()
```

```python
data["language"].value_counts()
```

```python
data=data.drop(columns=["language"])
```

## Explaratory data analysis

```python
data.describe()
```

```python
data.head()
```

```python
data.info()
```

```python
#N of unique values in each columns
data.nunique()
```

```python
data["availability"].unique()
```

```python
A=data.groupby("availability").count()["name"].to_frame("N Of products available").sort_values(by=["N Of products available"],ascending=False)
A
```

```python
plt.style.use("bmh")
plt.pie(A["N Of products available"],labels=A.index)
plt.title("Products available Instock and Outstock")
```

```python
B=data.groupby("color").count()["name"].to_frame("N of products available").sort_values(by=["N of products available"],ascending=False)
B
```

```python
sns.heatmap(B)
plt.title("Colors more available")
plt.style.use("classic")
```

```python
C=data.groupby("category").count()["name"].to_frame("N of products").sort_values(by=["N of products"],ascending=False)
C
```

```python
plt.pie(C["N of products"],labels=C.index,autopct="%1.1F%%")
```

```python
D=data.groupby("category").mean()[["average_rating","reviews_count"]]
D
```

```python
E=data.groupby("category").mean()["Discount"].to_frame("Discount").sort_values(by=["Discount"],ascending=False)
E
```

```python
E.plot(kind="bar")
plt.title("Adidas discount for each category")
plt.ylabel("Avg discount")
```

```python
F=data.groupby("availability").mean()["Discount"].to_frame("Discount").sort_values(by=["Discount"],ascending=False)
F
```

```python
F=data.groupby("color").mean()["reviews_count"].to_frame("reviews_count").sort_values(by=["reviews_count"],ascending=False)
F
```

```python
F.plot(kind="bar",color="g")
plt.title("Colors with highest number of reviews")
plt.xlabel("Color")
plt.ylabel("N of reviews")
plt.show()
```
